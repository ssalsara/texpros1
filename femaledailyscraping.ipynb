# BeautifulSoup

```python
from bs4 import BeautifulSoup
```

```python
import requests
page = requests.get('https://reviews.femaledaily.com/products/lips/lip-stain-tint?brand=&order=popular&page=1')
bs = BeautifulSoup(page.text, 'html.parser')
bs
```

```python
import json
item = bs.find_all('script', {'id': '__NEXT_DATA__'})
data = json.loads(item[0].text)

def extract_attributes(data):
    attributes = {}
    attributes['id'] = data['id']
    attributes['name'] = data['name']
    attributes['slug'] = data['slug']
    attributes['description'] = data['description']
    attributes['children'] = []
    if 'children' in data:
        for child in data['children']:
            attributes['children'].append(extract_attributes(child))
    return attributes

result = extract_attributes(data['props']['pageProps']['dehydratedState']['queries'][0]['state']['data']['data'])

print("Result:")
print(json.dumps(result, indent=4))
```

```python
import pandas as pd

# Extract the data from the result dictionary
data_list = []
def extract_data(attributes):
    data_list.append({
        'id': attributes['id'],
        'name': attributes['name'],
        'slug': attributes['slug'],
        'description': attributes['description']
    })
    if 'children' in attributes:
        for child in attributes['children']:
            extract_data(child)

extract_data(result)

# Create a Pandas DataFrame
df = pd.DataFrame(data_list)

# Print the resulting table
print(df)
```
